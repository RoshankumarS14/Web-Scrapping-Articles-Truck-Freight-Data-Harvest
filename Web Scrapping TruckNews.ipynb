```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
# Define the URL of the website
url = 'https://www.trucknews.com/news/page/'

page_headings = []

for page in range(1,2):

    # Send a GET request to the website
    response = requests.get(url+str(page)+"/")

    # Parse the HTML content of the page with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all headline elements
    # You may need to inspect the website and update the tag and class accordingly
    headlines = soup.find_all(['h2'])

    page_headings.append(headlines)
```

```python
list(pd.Series([a["href"] for a in soup.find_all(['a'])]).drop_duplicates())
```

```python
news_headlines = []
for headlines in page_headings:
    for headline in headlines:
        news_headlines.append(headline.text.strip())
```

```python
news_headlines = pd.DataFrame(news_headlines,columns=["Headlines"])
skip_words = ["Subscribe Centre","Read our digital edition","Topics","Contact","Sister Publications","Advertise with Us","Follow Us","Legal"]
news_headlines["Headlines"] = news_headlines["Headlines"].apply(lambda a : None if a in skip_words else a)
news_headlines.dropna(inplace=True)
news_headlines.to_csv("Truck News.csv",index=False)
news_headlines
```
